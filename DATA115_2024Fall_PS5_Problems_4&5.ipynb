```r
#Load
iris_df = read.csv("iris_data.csv")


#A
library(ggplot2)

x = "petal_length"
y = "petal_width"

ggplot(iris_df, aes(x= .data[[x]], y= .data[[y]], color = type)) +
geom_point(size = 3) +
labs(title = "Petal Width vs. Petal Length by species", x = "Petal Length", y = "Petal Width", color = "Species") +
theme_dark() + 
theme(plot.background = element_rect(fill = "grey"))

#B
four_numerical_columns = iris_df[, c("sepal_length", "sepal_width", "petal_length", "petal_width")]
iris_pca = prcomp(four_numerical_columns)
pca_df = data.frame(PC1 = iris_pca$x[, 1], PC2 = iris_pca$x[, 2], type = iris_df$type)

ggplot(pca_df, aes(x= PC1, y= PC2, color = type)) +
geom_point(size = 3) +
labs(title = "PCA Scatterplot", x = "Principle Component 1", y= "Principle Component 2", color = "Species") +
theme_dark() +
theme(plot.background = element_rect(fill = "grey"))

#C
summary(iris_pca)

#D
iris_pca$rotation
```
